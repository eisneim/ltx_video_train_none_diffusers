```python
import os, time, random
import torch
import matplotlib.pyplot as plt
import numpy as np

```

```python
source_dir = "/home/eisneim/www/ml/_video_gen/LTX-Video-3-22/data/mup"
sub_dir = "checkpoint-1500"

runs = {}

for ii in os.listdir(source_dir):
    data_file = os.path.join(source_dir, ii, sub_dir, "plot.pt")
    if os.path.exists(data_file):
        data = torch.load(data_file, weights_only=True)
        runs[ii] = data

print(runs.keys())
```

```python
plt.rcParams["figure.figsize"] = (10,5)
# fig = plt.figure(figsize=(20, 10))

m = 50
# grouped = 100
keys = list(runs.keys())
keys.sort(key=lambda x: float(x.split("_")[1]))

for key in keys:
    data = runs[key]
    y = np.array(data["losses"])
    x = np.arange(y.shape[0], dtype=y.dtype)
    # x /= m
    # plt.plot(x, np.log(y), label=f"train {key}", alpha=.5, linewidth=.05)

    n = y.shape[0] // m
    y = y[:n * m].reshape((n, m)).mean(axis=1)
    x = np.arange(n, dtype=y.dtype) + 0.5
    plt.plot(x, y, label=f"{key}", alpha=.8)
    
plt.xlabel(f"Epoch")
plt.ylabel("log loss")
plt.legend()
# loss_plot_path = os.path.join(self.args.output_dir, f"{name}.png")
# plt.savefig(loss_plot_path, bbox_inches="tight")
plt.show()
# plt.close()
```
